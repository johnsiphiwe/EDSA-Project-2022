```python
#libraries needed for webscrapping

import requests # request accesing pages
from bs4 import BeautifulSoup # to scrape html code from pages
import pandas as pd # to create a data frame
import numpy as np # generate random times for pauses between requests 
import re
import string
from datetime import datetime
```

We used indeed to search for data science jobs. search phrase = 'data scientist'

```python
https://za.indeed.com/jobs?q=data%20scientist&l=South%20Africa,%20Gauteng&start&vjk=431142e852804cc4
```

```python
def extract(page):
    #check my user agent
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36'}
    #use url from career site put page
    url = f'https://za.indeed.com/jobs?q=data%20scientist&l=South%20Africa,%20Gauteng&start={page}'
    res = requests.get(url, headers)
    #add soup variable
    soup = BeautifulSoup(res.content, 'lxml')
    return soup
```

```python
def transform(soup):
    articles = soup.find_all('div', class_ = 'job_seen_beacon')
    #return len(articles)
    #loop through all the articles
    for item in articles:
        try:
            title = item.find('h2', class_ ='jobTitle').text.strip()
            #title_ = title.find('span').text.strip()
        except:
            title = ''
        #company = item.find('a', class_ = 'turnstileLink companyOverviewLink').text.strip()
        company = item.find('a', class_ = 'turnstileLink companyOverviewLink')
        location = item.find('div', class_ = 'companyLocation').text.strip()
        try:
            salary = item.find('div', class_ = 'metadata salary-snippet-container').text.strip()
        except:
            salary = ''
        #print(salary)
        desc = item.find('div', class_ = 'job-snippet').text.strip().replace('\n', '')
        #print(desc)
        date = item.find('span', class_ = 'date').text.strip()
        i = item.find('tbody') #calling the table body to go inside of
        a = i.find('tr')
        #atag = item.find_all('a').text.strip()
        job_url = 'https://za.indeed.com' +str(a.find('href'))
        #job_url = 'https://www.indeed.com' +str(atag.get('href'))
        date_collected = datetime.today().strftime('%Y-%m-%d')
        
        jobs = {'title': title,
                'company': company,
                'location': location,
               'desc': desc,
               'date_posted': date, 
               'salary': salary,
                'job_link':job_url
               }
               #'job_link':job_url}
       
        job_list.append(jobs)
    return
```

```python
job_list = []
c = extract(0)
transform(c)
print(len(job_list))
```

```python
indeed_df = pd.DataFrame(job_list)
indeed_df
```
