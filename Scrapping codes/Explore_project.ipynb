```python
def clean_text(text):
    """This function is to clean the data removing urls, punctuations, spaces and making text to be lowercase """
    
    URL = r'http[s]?://(?:[A-Za-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9A-Fa-f][0-9A-Fa-f]))+'
    text = re.sub(URL, '', text)
    
    text = text.lower() # making text to be lowercase 
      
    text =re.sub("(\.)|(\;)|(\:)|(\!)|(\')|(\?)|(\,)|(\")|(\|)|(\()|(\))|(\[)|(\])|(\%)|(\$)|(\>)|(\<)|(\{)|(\})", "", text) # removing all punctuation with no space
    
    text =re.sub("(<br\s/><br\s/?)|(-)|(_)|(/)|(:).", " ", text) # removing all punctuation with a space
    
    text = re.sub("\\s+", " ", text)  # Remove extra whitespace
       
    text = re.sub(r"U+FFFD ", ' ', text) # Remove that funny diamond
    
    text = text.lstrip()  # removes whitespaces before string
    
    text = text.rstrip()  # removes whitespaces after string
    
    return text

df['job_desc_new'] = df['job_desc'].apply(clean_text)


# removing punctuations on our dataset
df['job_desc_new'] = df['job_desc_new'].apply(lambda x: ''.join([l for l in x if l not in string.punctuation]))
```

```python
#start by tokenizing the job description column
tokeniser = TreebankWordTokenizer()

# transforming the data using the Treebankword tokenizer
df['job_desc_new_tokenized'] = df['job_desc_new'].apply(tokeniser.tokenize)
```

```python
# converting the column job_desc_new in the dataframe into a list
texts = df['job_desc_new'].tolist()

# tagging the job description
tagged_texts = pos_tag_sents(map(word_tokenize, texts))
df['POS'] = tagged_texts

""" this function indentifies the parts of speech Noun, Adjective, Verb and Adverb """
def get_wordnet_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ

    elif tag.startswith('V'):
        return wordnet.VERB

    elif tag.startswith('N'):
        return wordnet.NOUN

    elif tag.startswith('R'):
        return wordnet.ADV
    
    else:
        return wordnet.NOUN
```

```python
# transforming the data with the function we created
df['POS'] = df['POS'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
df
```

### Lemmatize

```python
# initialising our lemmatizer
lemmatizer = WordNetLemmatizer()

# applying lemmatization to the POS column
df['job_desc_new'] = df['POS'].apply(lambda x: [lemmatizer.lemmatize(word, tag) for word, tag in x])

# joining the tokenised words after they have been lemmatized
df['job_desc_new'] = [' '.join(map(str, l)) for l in df['job_desc_new']]
```

### stopwords

```python
# Removing common words which appear in the column for description
remove_words = ['science','a', 'for', 'and', 'in', 'with', "we're", 'have', 'that', 'an', 'the', 'their', 'by', 'we', 'as', 'this', 'be', 'not', '000', 'g', 's', 'about']

# Function to remove common words listed above
def remove_common_words(message):
  pattern = re.compile(r'\b(' + r'|'.join(remove_words) + r')\b\s*')
  message = pattern.sub('', message)
  return message

df['word_desc'] = df['job_desc_new'].apply(remove_common_words)

# Adding select words to stop words for better analysis on important word frequency
stop = set(stopwords.words('english')) 
stop_words = ["I",'We','The'] + list(stop)

# Removing stop words from english language
df['word_desc'] = df['word_desc'].apply(lambda x: ' '.join([item for item in x.split() if item not in stop_words]))
```
