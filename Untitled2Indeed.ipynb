```python
#libraries needed for webscrapping

import requests # request accesing pages
from bs4 import BeautifulSoup # to scrape html code from pages
import pandas as pd # to create a data frame
import numpy as np # generate random times for pauses between requests 
import re
import string
from datetime import datetime
```

```python
def extract(page):
    #check my user agent
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36'}
    #use url from career site put page
    url = f'https://www.indeed.com/jobs?q=data%20science&start={page}'
    res = requests.get(url, headers)
    #add soup variable
    soup = BeautifulSoup(res.content, 'lxml')
    return soup
```

```python
def transform(soup):
    #jobs = soup.find_all('div', class_ ='slider_item')
    jobs = soup.find_all('div', class_ ='slider_item')
    for job in jobs:
        try:
            #title = job.find('h2', class_='jobTitle').get_text()
            title = job.find('a').text.strip()
        except:
            title = 'None'
        comp = soup.find('td', class_ ='resultContent')
        company_ = comp.find('div', class_ = 'heading6 company_location tapItem-gutter companyInfo')
        try:
            company = company_.find('span').text.strip()
        except:
            company = ''
       # try:
            #company = soup.find('span', class_ ='companyName')
        #except:
            #company = 'None'
        #try:
            #company_ = company.find('div', class_ = 'heading6 company_location tapItem-gutter companyInfo')
        #except:
            
        #co_name = company_.find('span').text.strip()
    
    #salary = soup.find('div', class_ = 'heading6 tapItem-gutter metadataContainer noJEMChips salaryOnly')
        try:
            salary = soup.find('div', 'metadata salary-snippet-container').text.strip()
        except:
            salary = ''
        try:
            location_ = soup.find('div', class_ ='companyLocation').text.strip()
        except:
            location_ = 'None'
            
        date_posted = soup.find('span', class_= 'date').text.strip()  ###
   
        desc = job.find('div', class_ = 'job-snippet').text.strip().replace('\n', '') ####
    #atag = item.find_all('a').text.strip()
        atag = soup.h2.a  ####
        
        job_url = 'https://www.indeed.com' +str(atag.get('href'))  ###
        date_collected = datetime.today().strftime('%Y-%m-%d')
#print(salary)
        jobs = {'title': title,
                'company': company,
                'location': location_,
               'job_desc': desc,
               'date_posted': date_posted,
                'date_scrapped': date_collected,
               'salary': salary,
               'job_url': job_url}
            
       
        job_list1.append(jobs)
    return
```
